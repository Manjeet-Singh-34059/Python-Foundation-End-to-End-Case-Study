### Dropping and Imputing Missing Values 

```python
essential_cols=[ "order_status", "order_purchase_timestamp", "customer_id", "product_id", 
                 "seller_id", "price", "payment_type", "payment_value"]
final_df=df.copy()
final_df=final_df.dropna(subset=essential_cols)

# Fill missing date/time fields
final_df["order_approved_at"]=final_df["order_approved_at"].fillna(final_df["order_purchase_timestamp"])
final_df["order_delivered_carrier_date"]=final_df["order_delivered_carrier_date"].ffill()
final_df["order_delivered_customer_date"]=final_df["order_delivered_customer_date"].ffill()

# Fill missing product/seller details
final_df["product_category_name"]=final_df["product_category_name"].fillna("unknown")
final_df["product_name_lenght"]=final_df["product_name_lenght"].fillna(0)
final_df["product_description_lenght"]=final_df["product_description_lenght"].fillna(0)
final_df["product_photos_qty"]=final_df["product_photos_qty"].fillna(0)
final_df["product_weight_g"]=final_df["product_weight_g"].fillna(final_df["product_weight_g"].median())
final_df["product_length_cm"]=final_df["product_length_cm"].fillna(final_df["product_length_cm"].median())
final_df["product_height_cm"]=final_df["product_height_cm"].fillna(final_df["product_height_cm"].median())
final_df["product_width_cm"]=final_df["product_width_cm"].fillna(final_df["product_width_cm"].median())

# Fill missing seller location data
final_df["seller_city"]=final_df["seller_city"].fillna("unknown")
final_df["seller_state"]=final_df["seller_state"].fillna("unknown")
```

### c. Understand the retention of customers on month on month basis

```python
# creating order_month from order_purchase_timestamp
final_df["order_month"]=final_df["order_purchase_timestamp"].dt.to_period("M").astype("str")

# unique customer per month
monthly_customer=final_df.groupby("order_month")["customer_unique_id"].apply(set).reset_index()
```

```python
# Create a new DataFrame to store retention
retention_list=[]

for i in range(1,len(monthly_customer)):
    prev_month=monthly_customer.loc[i-1,"customer_unique_id"]
    curr_month=monthly_customer.loc[i,"customer_unique_id"]

    retained_customer=prev_month.intersection(curr_month)

    retention_rate= len(retained_customer) / len(prev_month) if len(prev_month) > 0 else 0

    retention_list.append({  
        "Month": monthly_customer.loc[i,"order_month"],
        "Customer_Last_Month": len(prev_month),
        "Retained_Customer" : len(retained_customer),
        "Retention_Rate" : round(retention_rate,4)
    })

# Creation Retention DataFrame
retention_df=pd.DataFrame(retention_list)
```

# 3. Cross-Selling (Which products are selling together)
### Hint: We need to find which of the top 10 combinations of products are selling together in each transaction. (combination of 2 or 3 buying together)


```python
# Use only needed columns
cross_df=final_df[["order_id","product_id"]]

# Group product IDs under each order
order_products=cross_df.groupby("order_id")["product_id"].apply(list).reset_index()
```

```python
from itertools import combinations
from collections import Counter

combo_counter=Counter()

# Count combinations per order
for products in order_products["product_id"]:
    # Remove duplicates within the same order (optional)
    unique_products = list(set(products))

    # Generate combos of size 2 and 3
    for r in [2,3]:
        if len(unique_products)>=r:
            combos=combinations(sorted(unique_products), r)
            combo_counter.update(combos)
```

```python
# Convert to DataFrame
combo_df=pd.DataFrame(combo_counter.items(), columns=["Product_Combo", "Count"])

# Sort and get top 10
top_combos=combo_df.sort_values("Count", ascending=False).head(10)

top_combos.reset_index(drop=True, inplace=True)
top_combos
```

```python
top_combos = top_combos.sort_values("Count", ascending=True)
top_combos["Product_Combo"] = top_combos["Product_Combo"].astype(str)  

plt.figure(figsize=(12, 6))
sns.barplot(data=top_combos, y="Product_Combo", x="Count",hue="Product_Combo",palette="viridis")
plt.title("Top 10 Product Combinations (Cross-Selling)")
plt.xlabel("Number of Orders")
plt.ylabel("Product Combination")
plt.tight_layout()
plt.show()
```
